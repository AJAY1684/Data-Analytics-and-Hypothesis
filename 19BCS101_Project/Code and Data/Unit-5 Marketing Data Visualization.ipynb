```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline                  
# to dipaly graph in jupyter
```

```python
Marketing = pd.read_csv("Company_Data.csv",index_col=0)
```

```python
# Education of Customers who purchased from Company

plot = sns.countplot(x = "Education", data = Marketing)
plot.set_xticklabels(plot.get_xticklabels(), rotation=40);
```

```python
# Country of Customers who purchased from Company

plot = sns.countplot(x = "Country", data = Marketing)
plot.set_xticklabels(plot.get_xticklabels(), rotation=40);
```

```python
# Analyze the spread of the "Income"

sns.distplot(Marketing["Age"], kde=False); 
```

```python
# Analyze the spread of the "Income"

sns.distplot(Marketing["Income"], kde=False);  
```

# Analysing Relationship between Amount spent on each product and total money Money spent

```python
sf_cols = [col_name for col_name in Marketing.columns if "Mnt" in col_name] #this is list comprehension


fig, axs = plt.subplots(nrows = 3,ncols = 3, figsize = (10,10))
for i in range(0, len(sf_cols)):
    rows = i // 3
    cols = i % 3
    ax = axs[rows, cols]
    plot = sns.regplot(x = sf_cols[i], y = "Purchases", data = Marketing, ax=ax)    
```
